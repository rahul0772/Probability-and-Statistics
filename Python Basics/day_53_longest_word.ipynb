```python
# PROBLEM:
# --------
# Given a sentence, find the longest word in that sentence.
#
# Example:
# Input  : "I love learning Python programming"
# Output : "programming"
#
# But we will solve this in a way that teaches you
# CORE PYTHON deeply and clearly.
#
# ===============================================================



# ---------------------------------------------------------------
# STEP 1: DEFINE A FUNCTION
# ---------------------------------------------------------------
# A function is like a machine.
# You give it something (input).
# It works.
# It returns something (output).
#
# Syntax:
# def function_name(parameters):
#     code
#
# "def" means DEFINE
# ---------------------------------------------------------------

def find_longest_word(sentence):

    # -----------------------------------------------------------
    # STEP 2: UNDERSTAND WHAT "sentence" IS
    # -----------------------------------------------------------
    # sentence is a STRING.
    #
    # A string is a collection of characters.
    # Example:
    # "Hello"
    # H = character
    # e = character
    #
    # Strings are inside quotes.
    # -----------------------------------------------------------


    # -----------------------------------------------------------
    # STEP 3: SPLIT THE SENTENCE INTO WORDS
    # -----------------------------------------------------------
    # sentence.split()
    #
    # What does split() do?
    # It breaks the string into pieces.
    #
    # Example:
    # "I love Python"
    #
    # After split():
    # ["I", "love", "Python"]
    #
    # Now it becomes a LIST.
    #
    # What is a LIST?
    # A list is a collection of items.
    # It uses square brackets [].
    # -----------------------------------------------------------

    words = sentence.split()

    # Now words is a LIST.
    # Example:
    # ["I", "love", "learning", "Python"]


    # -----------------------------------------------------------
    # STEP 4: CREATE VARIABLES TO TRACK LONGEST WORD
    # -----------------------------------------------------------

    longest_word = ""
    # We start with empty string ""
    # Why?
    # Because we haven't found any word yet.

    max_length = 0
    # We start with 0 because no word length checked yet.


    # -----------------------------------------------------------
    # STEP 5: LOOP THROUGH EACH WORD
    # -----------------------------------------------------------
    #
    # A loop repeats code.
    #
    # for word in words:
    # This means:
    # Take one word at a time from the list.
    #
    # Example:
    # If words = ["I", "love", "Python"]
    #
    # First loop:
    # word = "I"
    #
    # Second loop:
    # word = "love"
    #
    # Third loop:
    # word = "Python"
    #
    # -----------------------------------------------------------

    for word in words:

        # -------------------------------------------------------
        # STEP 6: FIND LENGTH OF CURRENT WORD
        # -------------------------------------------------------
        #
        # len(word) gives number of characters.
        #
        # Example:
        # len("cat") = 3
        # len("Python") = 6
        #
        # -------------------------------------------------------

        current_length = len(word)


        # -------------------------------------------------------
        # STEP 7: COMPARE LENGTHS
        # -------------------------------------------------------
        #
        # We check:
        # Is this word longer than previous longest?
        #
        # if condition:
        #
        # Example:
        # if 5 > 3:
        #     True
        #
        # -------------------------------------------------------

        if current_length > max_length:

            # If this word is longer,
            # we update both values.

            max_length = current_length
            # Now this becomes the new biggest length.

            longest_word = word
            # Now this becomes the new longest word.


    # -----------------------------------------------------------
    # STEP 8: RETURN RESULT
    # -----------------------------------------------------------
    #
    # return means:
    # Send this value back to whoever called the function.
    #
    # -----------------------------------------------------------

    return longest_word




# ===============================================================
# TESTING OUR FUNCTION
# ===============================================================

# We give input sentence here.
sentence_input = "I love learning Python programming because it is powerful"

# Call the function
result = find_longest_word(sentence_input)

# Print result
print("The longest word is:", result)



# ===============================================================
# üß† IMPORTANT CORE CONCEPTS YOU JUST LEARNED
# ===============================================================
#
# 1Ô∏è‚É£ VARIABLES
#    longest_word = ""
#    max_length = 0
#
# 2Ô∏è‚É£ STRING
#    "Hello"
#
# 3Ô∏è‚É£ LIST
#    ["I", "love", "Python"]
#
# 4Ô∏è‚É£ LOOP
#    for word in words:
#
# 5Ô∏è‚É£ CONDITION
#    if current_length > max_length:
#
# 6Ô∏è‚É£ FUNCTION
#    def find_longest_word(sentence):
#
# 7Ô∏è‚É£ TIME COMPLEXITY (Expert Thinking)
#
#    If sentence has N words,
#    We check each word ONCE.
#
#    So time complexity is:
#    O(N)
#
#    That means:
#    If words double,
#    time doubles.
```
