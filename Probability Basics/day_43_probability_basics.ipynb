## 1. **What is a Random Variable?**

A **random variable** is like a **mystery box** where you donâ€™t know what the result will be. Itâ€™s something that can take different values based on **chance**.

Think about a **coin flip**. The result can be **Heads** or **Tails**, but you donâ€™t know what will happen until you flip the coin. The result (Heads or Tails) is a **random variable**.

### Example: Coin Flip ðŸª™

- If you flip a coin, the **random variable** is the **result** of the flip. It can either be **Heads** or **Tails**.
- **Heads** and **Tails** are the possible values the random variable can take.
- You canâ€™t predict the result exactly, but you know it will be either Heads or Tails. That's why it's called a **random variable**.

So, a **random variable** is just something that can **change** because of **chance**.

## 2. **What is a Discrete Random Variable?**

A **discrete random variable** is a special type of **random variable** that can only take **specific** or **countable** values.

### Example: Rolling a Dice ðŸŽ²

- When you roll a **dice**, the **random variable** is the **number** that shows up on the dice.
- The dice has **6 sides**, so it can land on one of **6 numbers**: 1, 2, 3, 4, 5, or 6.
- You **cannot** roll a 2.5 or 7 on a dice! The numbers are **discrete** because they are countable, and there are no numbers in between.

### Why "Discrete"?

- **Discrete** means **separate**, like counting fingers or apples. You canâ€™t have **half** an apple in this case â€” you count **whole apples**. So, a **discrete random variable** has a fixed set of possible outcomes.

- The dice has 6 possible outcomes: **1, 2, 3, 4, 5, and 6**. We can list them and count them.

## 3. **What is a Probability Mass Function (PMF)?**

A **Probability Mass Function (PMF)** is just a **fancy table** that shows the **chance** of each possible outcome for a **discrete random variable**.

Imagine you're rolling a dice again. The PMF will show the probability (chance) for each of the 6 numbers you might roll.

### Example: Rolling a Dice ðŸŽ² (Again!)

Letâ€™s say we are rolling a **fair** dice. This means that every number on the dice has the **same chance** of happening.

The **PMF** for this dice roll looks like this:

| **Outcome (What We Roll)** | **Probability (Chance)** |
|---------------------------|-------------------------|
| 1                         | 1/6                     |
| 2                         | 1/6                     |
| 3                         | 1/6                     |
| 4                         | 1/6                     |
| 5                         | 1/6                     |
| 6                         | 1/6                     |

### Explanation:
- When you roll the dice, the chance of getting any **one number** (like 1, 2, 3, etc.) is the same: **1 out of 6**.
- So, each number has a **1/6 chance** (or a **16.7% chance**) of showing up.
- **The PMF** is just a **fancy name** for showing the chances for each possible result.

## 4. **What is Bayes' Theorem?**

Now letâ€™s talk about **Bayesâ€™ Theorem**. This is a way to **update** your guess based on new information. It helps you find the **chance** of something happening after you learn something new!

### Donâ€™t Worry About The Formula!

Bayesâ€™ Theorem might look a bit complicated, but Iâ€™ll explain it in **super simple terms** so it makes sense. ðŸŒŸ

### Example: Picking a Red Ball from a Bag ðŸŽ±

Imagine you have a bag with **5 balls**: **3 red balls** and **2 green balls**.

You **close your eyes**, pick a ball, and when you open your eyes, you see that itâ€™s **red**. Now, you want to know: **Whatâ€™s the chance** that the ball you picked is **one of the red balls** from the bag?

### Step 1: Understand the Situation

- You know there are **5 balls in total**, and **3 are red**, so the chance of picking a red ball is **3 out of 5** (or **3/5**).
  
- The chance of picking a **green ball** is **2 out of 5** (or **2/5**).

Now, you already know you picked a **red ball**.

### Step 2: Use Bayes' Theorem to Update Your Guess

Bayesâ€™ Theorem is just a way of saying: **If I know that I picked a red ball, how sure am I that itâ€™s one of the 3 red balls in the bag?**

#### Formula (Donâ€™t worry about this part!):

p(A|B) = p(B|A) * p(A) / p(B)

- **p(A|B)** = What we want to find: the chance that the ball is **red**, given that you picked a **red** ball.
- **p(B|A)** = If the ball is **red**, then you **definitely** picked a red ball. So this is 100% or **1**.
- **p(A)** = The chance of picking any **red** ball (which is 3 out of 5 or **3/5**).
- **p(B)** = The chance of picking a **red** ball (which is also 3 out of 5 or **3/5**).

#### Step 3: Putting it Together

Letâ€™s plug the numbers into the formula:

p(A|B) = (1 * 3/5) / (3/5)

p(A|B) = **1**

### What does this mean?

After you know the ball is **red**, you are **100% sure** that itâ€™s one of the **3 red balls** in the bag!

So, **Bayes' Theorem** is a way to update your guess when you get new information. In this case, once you know you picked a red ball, youâ€™re sure itâ€™s one of the red balls in the bag.

## **Letâ€™s Recap Everything:**

1. **Random Variable**: Something that can **change** based on **chance**. For example, the result of flipping a coin (Heads or Tails) is a random variable.
2. **Discrete Random Variable**: A type of random variable that can only take **certain, countable** values. For example, when you roll a dice, the possible values are **1, 2, 3, 4, 5, 6**.
3. **PMF (Probability Mass Function)**: A table that shows the **probabilities** of each possible value for a discrete random variable. For example, the chance of rolling a 1, 2, 3, etc., on a fair dice is 1/6.
4. **Bayes' Theorem**: A way to update your guess based on new information. For example, once you know you picked a red ball, youâ€™re 100% sure itâ€™s one of the red balls in the bag!

## **Even Simpler Example of Bayes' Theorem:**

Imagine you have a **bag with 2 red balls** and **3 green balls**. So there are **5 balls** in total.

You pick a ball, and itâ€™s **green**.

- The probability of picking a **green** ball is **3 out of 5** (3 green balls out of 5 total balls).

### After learning the ball is green, Bayes' Theorem helps you be **sure** it is one of the **green balls** in the bag!

```python
# Let's start by importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# First, let's deal with **Random Variables** and **Discrete Random Variables**

# Imagine you are rolling a fair dice.
# The random variable here is the number you get when you roll the dice.

# We'll create a list to represent all possible outcomes of a dice roll.
outcomes = [1, 2, 3, 4, 5, 6]

# Now, let's simulate a dice roll
# We'll randomly choose one of these outcomes (this is our "random variable" in action)
dice_roll = np.random.choice(outcomes)
print(f"Simulated dice roll result: {dice_roll}")

# Now let's make a **Probability Mass Function (PMF) for this dice roll**.

# Since this is a fair dice, each number has the same probability of happening,
# so the probability for each outcome is 1/6.

# Create a list of probabilities for each outcome (1, 2, 3, 4, 5, 6).
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Plot the PMF (it will show the probability of each dice outcome)
plt.figure(figsize=(8,6))
plt.bar(outcomes, probabilities, color='#80EF80')
plt.title("Probability Mass Function for a Fair Dice Roll")
plt.xlabel("Dice Outcome")
plt.ylabel("Probability")
plt.xticks(outcomes)
plt.show()

# **Bayes' Theorem Example: Let's say we have a bag with 3 red balls and 2 green balls.**

# We want to calculate the probability of picking a red ball, given that we know we picked a red ball.
# Bayes' Theorem formula: P(A|B) = P(B|A) * P(A) / P(B)

# Let's define the values:
# P(A) = The chance of picking a red ball, which is 3 red balls out of 5 total balls, so P(A) = 3/5
# P(B|A) = If the ball is red, we will definitely pick a red ball, so P(B|A) = 1
# P(B) = The total chance of picking any red ball, which is also 3 red balls out of 5 total balls, so P(B) = 3/5

P_A = 3 / 5  # Probability of picking a red ball
P_B_given_A = 1  # If the ball is red, we definitely picked a red ball
P_B = 3 / 5  # Total probability of picking a red ball

# Using Bayes' Theorem to calculate P(A|B) (the probability of picking a red ball after knowing it is red)
P_A_given_B = (P_B_given_A * P_A) / P_B
print(f"The probability of picking a red ball, given that it is red (P(A|B)): {P_A_given_B}")

# This should output 1, because if you know you picked a red ball, you're 100% sure it was one of the red balls.

# Letâ€™s visualize the bag with a pie chart to understand the red and green balls better:

labels = ['Red Balls', 'Green Balls']
sizes = [3, 2]  # 3 red balls, 2 green balls
colors = ['#FF746C', '#80EF80']

# Create a pie chart to represent the bag of balls
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Bag of Balls (Red and Green)")
plt.show()

# **Let's simulate a few dice rolls and apply the PMF and Bayes' Theorem together:**

# Simulate rolling the dice 10 times and keep track of the results
dice_rolls = np.random.choice(outcomes, size=10)  # simulate 10 rolls of the dice
print("Results of 10 dice rolls:", dice_rolls)

# Count how many times each number appears (we will count the occurrences of 1, 2, 3, 4, 5, 6)
counts = [list(dice_rolls).count(i) for i in outcomes]

# Now let's visualize how often each number came up during the 10 rolls
plt.figure(figsize=(8,6))
plt.bar(outcomes, counts, color='lightcoral')
plt.title("Dice Rolls (Frequency of Each Outcome)")
plt.xlabel("Dice Outcome")
plt.ylabel("Frequency")
plt.xticks(outcomes)
plt.show()

# **Final part - Combine everything into a fun experiment:**

# Let's simulate a few cases of Bayes' Theorem and dice rolls combined together:

# Let's say we know that we rolled a 4 on the dice.
# We now want to know the chance that the dice roll was a 4 given that we have rolled an even number (2, 4, or 6).

# P(A) = The chance of rolling a 4 = 1/6
# P(B|A) = If the number is 4, we will definitely pick a 4
# P(B) = The total chance of rolling an even number (2, 4, or 6) = 3/6 = 1/2

P_A = 1 / 6  # Probability of rolling a 4
P_B_given_A = 1  # If the dice is 4, you will definitely pick a 4
P_B = 3 / 6  # Probability of rolling any even number (2, 4, or 6)

# Applying Bayes' Theorem
P_A_given_B = (P_B_given_A * P_A) / P_B
print(f"The probability of rolling a 4, given that you rolled an even number: {P_A_given_B}")

# This shows us how **Bayes' Theorem**
# helps us update our beliefs based on new information (in this case, knowing it was an even number).
```
