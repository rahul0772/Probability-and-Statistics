# Easy Explanation of Probability and Counting Concepts

## 1. **Counting: Permutations and Combinations**

### **Permutations (Order Matters!)**

When you arrange things and **order** is important, it's called **permutations**. Think of it like arranging things in a line, where the order of everything makes a difference.

**Example 1:**
Imagine you have 3 toys: a **car**, a **doll**, and a **ball**. You want to know how many ways you can line them up on a shelf. The order is **important**!

- You can line them up in **6 different ways**:
  1. Car, Doll, Ball
  2. Car, Ball, Doll
  3. Doll, Car, Ball
  4. Doll, Ball, Car
  5. Ball, Car, Doll
  6. Ball, Doll, Car

So, there are **6 ways** to arrange them. This is what we call a **permutation**.

---

### **Combinations (Order Doesn't Matter!)**

Combinations are when you choose things, but **the order doesn’t matter**. It's like picking toys to play with, but it doesn’t matter if you pick the car first or the doll first—you just care about what toys you picked, not the order.

**Example 2:**
You have 3 toys again: a **car**, a **doll**, and a **ball**. Now, you're going to pick **2 toys** to play with, and you don’t care about the order!

- You can choose the toys in **3 ways**:
  1. Car and Doll
  2. Car and Ball
  3. Doll and Ball

So, there are **3 ways** to choose 2 toys. The order of the toys doesn’t matter, which makes it a **combination**.

---

## 2. **Probability Axioms (Rules of Probability)**

### **Rule 1: Probability is Between 0 and 1**

- The probability of something that **won’t** happen is **0** (like getting a 7 on a 6-sided die).
- The probability of something that **will** happen is **1** (like the sun coming up tomorrow).

**Example 3:**
When you flip a coin:
- The probability of **Heads** is **1/2** (or 0.5).
- The probability of **Tails** is also **1/2**.

This means there’s an equal chance of getting **Heads** or **Tails**. Both are between 0 and 1!

---

### **Rule 2: The Probability of Everything Happening is 1**

- If you list **all possible outcomes** of something, the chance of **something** happening is always **1**.

**Example 4:**
When you roll a die, the possible outcomes are **{1, 2, 3, 4, 5, 6}**.
- The probability of rolling one of these numbers is:
  - **1/6** for 1, **1/6** for 2, **1/6** for 3, and so on.
- **1/6 + 1/6 + 1/6 + 1/6 + 1/6 + 1/6 = 1**
- This means the probability of **any number** on the die is **1**, because **one of them** has to happen.

---

### **Rule 3: Adding Probabilities for "Or" Events**

If you're trying to find the probability of **either one event or another** happening, you **add** the probabilities. But be careful: if the events **overlap**, you subtract the overlap.

**Example 5:**
You roll a die and want the probability of getting either a **1** or a **2**.
- P(1) = **1/6**
- P(2) = **1/6**
- To find the probability of getting **1 or 2**:
  - P(1 or 2) = P(1) + P(2) = **1/6 + 1/6 = 2/6 = 1/3**
  
So, the probability of getting **either a 1 or a 2** is **1/3**.

---

## 3. **Sample Space (All Possible Outcomes)**

The **sample space** is just a fancy name for **all the possible things** that could happen in an event.

**Example 6:**
- When you flip a coin, the sample space is **{Heads, Tails}** because those are the only two things that can happen.
- When you roll a die, the sample space is **{1, 2, 3, 4, 5, 6}** because those are the only possible numbers you can roll.

---

## 4. **Events (What Happens)**

An **event** is just a thing that can happen. It’s like the outcome you're looking for.

**Example 7:**
- When you flip a coin, an event might be getting **Heads**. The event would be **{Heads}**.
- If you roll a die and want the event of **getting a number greater than 4**, your event would be **{5, 6}**.

---

## 5. **Independent Events (Events That Don't Affect Each Other)**

Two events are **independent** if what happens in one event doesn’t change what happens in the other event.

**Example 8:**
- Imagine you flip a coin and roll a die. The result of the coin flip (Heads or Tails) doesn’t change the die roll.
- The **coin flip** and **die roll** are **independent** events because one doesn’t affect the other.

**Probability Example:**
- Probability of getting **Heads** on a coin flip = **1/2**
- Probability of getting **6** on a die roll = **1/6**

To find the **probability of both** happening (getting Heads **and** rolling a 6), you multiply the probabilities:
- P(Heads and 6) = P(Heads) × P(6) = **1/2 × 1/6 = 1/12**

So, the chance of both happening is **1/12**.

---

## 6. **Mutually Exclusive Events (Events That Can’t Happen Together)**

Two events are **mutually exclusive** if they can’t happen at the same time.

**Example 9:**
- If you flip a coin, **getting Heads** and **getting Tails** are mutually exclusive. You can’t get both at once. You can only get one or the other.
  
**Example 10:**
- When you roll a die, **getting an odd number** (1, 3, 5) and **getting an even number** (2, 4, 6) are mutually exclusive. You can’t get both at the same time.

---

   - **Probability is between 0 and 1** (from impossible to certain).
   - The **probability of everything happening is 1**.
   - **Adding probabilities for "or" events**.
4. **Sample Space**: All the possible outcomes (like flipping a coin: {Heads, Tails}).

## Extra Fun Examples (For Better Understanding!):

**Example 11:**
Imagine you have a box with **3 red balls**, **2 blue balls**, and **1 green ball**. If you randomly pick one ball from the box, that’s an event. The **sample space** is **{red, red, red, blue, blue, green}**.

**Example 12:**
If you want the event of picking a **red ball**, the **probability** is:
- P(Red) = **3/6** = **1/2** (because there are 3 red balls and 6 balls in total).



```python
# First, let's import the math library so we can use functions like factorial and combinations.
import math
import random

# --------------------------------------------------------------
# 1. Permutations (Order Matters)
# Example: Let's say you have 3 toys: a car, a doll, and a ball.
# How many different ways can you arrange them in a line?
# Order matters in permutations.

toys = ["car", "doll", "ball"]  # Our toys are car, doll, and ball

# Let's count how many ways we can arrange them.
# The number of ways to arrange n items is called a "permutation" and is given by n! (n factorial).
# For 3 toys, we will calculate 3! which is 3 * 2 * 1 = 6.

num_toys = len(toys)  # Count how many toys we have (3 toys)

# Now we calculate how many ways we can arrange these toys using factorial.
permutations = math.factorial(num_toys)  # math.factorial does the math for us (3! = 6)

print("Permutations (how many ways to arrange 3 toys):")
print(permutations)  # Output will be 6, because 3! = 6

# --------------------------------------------------------------
# 2. Combinations (Order Doesn't Matter)
# Now, let's say we want to know how many ways we can pick 2 toys out of the 3 toys.
# But **order doesn't matter** in combinations. You just care about which toys you picked.

# Formula for combinations is C(n, r) = n! / (r! * (n - r)!)
# Where n = 3 (the total toys), r = 2 (how many toys we want to pick)

combinations = math.comb(num_toys, 2)  # math.comb calculates the number of ways to choose 2 toys from 3

print("\nCombinations (how many ways to choose 2 toys out of 3):")
print(combinations)  # Output will be 3, because there are 3 ways to choose 2 toys from 3

# --------------------------------------------------------------
# 3. Probability Axioms (Rules of Probability)
# Let's calculate the probability of rolling a 1 on a 6-sided die.

# A die has 6 faces, so the total number of outcomes is 6.
# The probability of getting a 1 is 1 out of 6, which is 1/6.
prob_1 = 1 / 6  # Probability of getting a 1 on a die is 1 out of 6

print("\nProbability of rolling a 1 on a 6-sided die:")
print(prob_1)  # This should be 0.1667 (1/6)

# --------------------------------------------------------------
# 4. Sample Space
# The **sample space** is all the possible outcomes of an event.
# For example, when you flip a coin, the possible outcomes are Heads or Tails.

coin_flip = random.choice(["Heads", "Tails"])  # We simulate a coin flip, randomly choosing between Heads or Tails

sample_space = ["Heads", "Tails"]  # The sample space for a coin flip is Heads or Tails
print("\nSample space for a coin flip:")
print(sample_space)  # This should print ['Heads', 'Tails']

print("\nCoin flip result:")
print(coin_flip)  # This will print either Heads or Tails, depending on the random choice

# --------------------------------------------------------------
# 5. Events
# An **event** is something that can happen. For example, let's simulate a die roll and check if the event "getting a 3" happened.

roll = random.randint(1, 6)  # Simulate a die roll by choosing a random number between 1 and 6
event_3 = roll == 3  # Check if the roll is a 3 (this will be True if the roll is 3, False otherwise)

print("\nRolling the die... Result:")
print(roll)  # This will print the result of the die roll (a number between 1 and 6)

print("\nDid we roll a 3?")
print(event_3)  # This will print True if we rolled a 3, False if we rolled a different number

# --------------------------------------------------------------
# 6. Independent Events
# Two events are **independent** if the outcome of one doesn't affect the other.
# For example, flipping a coin and rolling a die are independent events.

# We'll flip a coin and roll a die, and show that the two events don't affect each other.

coin_result = random.choice(["Heads", "Tails"])  # Simulate a coin flip
die_roll = random.randint(1, 6)  # Simulate a die roll

print("\nCoin flip result:")
print(coin_result)  # This will print either Heads or Tails

print("\nDie roll result:")
print(die_roll)  # This will print a number between 1 and 6

# Probability of getting Heads on a coin flip is 1/2, and the probability of rolling a 6 on a die is 1/6.
# The probability of both happening is the product of these probabilities.

prob_independent = (1 / 2) * (1 / 6)  # Probability of getting Heads AND rolling a 6 is (1/2) * (1/6)

print("\nProbability of flipping Heads and rolling a 6:")
print(prob_independent)  # This will print 0.0833 (1/12), the probability of both events happening

# --------------------------------------------------------------
# 7. Mutually Exclusive Events
# Two events are **mutually exclusive** if they cannot happen at the same time.
# For example, you can't get both Heads and Tails on a single coin flip.

# Let's simulate two coin flips and see if we can get both Heads and Tails at the same time.

flip1 = random.choice(["Heads", "Tails"])  # First coin flip
flip2 = random.choice(["Heads", "Tails"])  # Second coin flip

print("\nFirst flip result:")
print(flip1)  # This will print either Heads or Tails

print("\nSecond flip result:")
print(flip2)  # This will print either Heads or Tails

# Check if the two flips are mutually exclusive (they cannot be both Heads and Tails at once)
print("\nAre both flips Heads and Tails?")
print(flip1 != flip2)  # This will print True because we cannot get both Heads and Tails in one flip
```

Probability is the chance of something happening. It’s usually written as:
- A **fraction** (e.g., 3/5),
- A **decimal** (e.g., 0.6),
- Or a **percentage** (e.g., 60%).

---

## **1. Marginal Probability** (Chance of One Event)

### What is Marginal Probability?

Marginal probability is the chance of **one event** happening by itself.

### Example 1:

Imagine you have a bag with **5 balls**:
- 3 red balls
- 2 blue balls

We want to know: **What is the chance of picking a red ball?**

1. There are **3 red balls** and **2 blue balls**, so the total number of balls is **5**.
2. **Marginal Probability** is the chance of picking a red ball, which we can write as:

   **Chance of Red Ball = (Number of Red Balls) ÷ (Total Number of Balls)**

   - There are **3 red balls**.
   - There are **5 total balls** in the bag.

So, the fraction would be:

**3 red balls ÷ 5 total balls = 3/5**.

The chance of picking a red ball is **3 out of 5**, which is **60%** or **0.6**.

---

### Example 2:

Now, let’s use a new bag with **4 green balls** and **1 yellow ball** (so 5 balls total).

We want to know: **What is the chance of picking a green ball?**

1. There are **4 green balls** and **1 yellow ball**, so the total number of balls is **5**.
2. **Marginal Probability** for picking a green ball is:

**Chance of Green Ball = (Number of Green Balls) ÷ (Total Number of Balls)**

- There are **4 green balls**.
- There are **5 total balls** in the bag.

So, the fraction would be:

**4 green balls ÷ 5 total balls = 4/5**.

The chance of picking a green ball is **4 out of 5**, which is **80%** or **0.8**.

---

## **2. Conditional Probability** (If You Know Something Already)

### What is Conditional Probability?

Conditional probability is the chance of something happening **after** you already know that something else has happened.

### Example 1:

Let’s go back to the bag with **5 balls** (3 red, 2 blue). This time, **you already picked a red ball**. We want to know: **What is the chance that the next ball you pick will also be red?**

1. After picking a red ball, there are **2 red balls left** and **4 balls total**.
2. The **conditional probability** of picking another red ball is:

**Chance of Red Ball after Picking One = (Number of Red Balls Left) ÷ (Total Balls Left)**

- There are **2 red balls left**.
- There are **4 balls left in total**.

So, the fraction would be:

**2 red balls ÷ 4 total balls = 2/4**.

The chance of picking another red ball is **2 out of 4**, which is **50%** or **0.5**.

---

### Example 2:

Now, let’s use a new bag with **4 red balls** and **2 blue balls** (6 balls total).

If you **already picked a blue ball**, we want to know: **What is the chance that the next ball you pick will be red?**

1. After picking the blue ball, there are **4 red balls left** and **5 balls total**.
2. The **conditional probability** of picking a red ball is:

**Chance of Red Ball after Blue = (Number of Red Balls Left) ÷ (Total Balls Left)**

- There are **4 red balls left**.
- There are **5 balls left in total**.

So, the fraction would be:

**4 red balls ÷ 5 total balls = 4/5**.

The chance of picking a red ball is **4 out of 5**, which is **80%** or **0.8**.

---

## **3. Joint Probability** (Two Things Happening Together)

### What is Joint Probability?

Joint probability is the chance of **two things happening together** (in sequence).

### Example 1: Picking Two Balls in Sequence

Let’s go back to the bag with **5 balls** (3 red, 2 blue). We want to know: **What is the chance of picking a red ball first and a blue ball second?**

#### Step 1: Find the chance of picking a red ball first.

- The total number of balls is **5**.
- The number of red balls is **3**.

So, the chance of picking a red ball first is:

**Chance of Red Ball First = (Number of Red Balls) ÷ (Total Number of Balls) = 3/5**.

This is **3 out of 5**, or **60%**.

#### Step 2: After picking the red ball, find the chance of picking a blue ball second.

- After picking the red ball, there are **2 blue balls left** and **4 balls total**.

So, the chance of picking a blue ball second is:

**Chance of Blue Ball Second = (Number of Blue Balls Left) ÷ (Total Balls Left) = 2/4**.

This is **2 out of 4**, or **50%**.

#### Step 3: Multiply the two probabilities to get the joint probability.

To find the **joint probability** of both events happening together, we multiply the two probabilities:

**P(Red First and Blue Second) = (3/5) × (2/4) = 6/20 = 0.3 (or 30%)**

**In simple words**: There’s a **30% chance** that you’ll pick a red ball first and a blue ball second.

---

### Example 2: Picking Two Balls in a Different Order

Let’s use the same bag with **3 red balls** and **2 blue balls**. We want to know: **What is the chance of picking a blue ball first and a red ball second?**

#### Step 1: Find the chance of picking a blue ball first.

- The total number of balls is **5**.
- The number of blue balls is **2**.

So, the chance of picking a blue ball first is:

**Chance of Blue Ball First = (Number of Blue Balls) ÷ (Total Number of Balls) = 2/5**.

This is **2 out of 5**, or **40%**.

#### Step 2: After picking the blue ball, find the chance of picking a red ball second.

- After picking the blue ball, there are **3 red balls left** and **4 balls total**.

So, the chance of picking a red ball second is:

**Chance of Red Ball Second = (Number of Red Balls Left) ÷ (Total Balls Left) = 3/4**.

This is **3 out of 4**, or **75%**.

#### Step 3: Multiply the two probabilities to get the joint probability.

**P(Blue First and Red Second) = (2/5) × (3/4) = 6/20 = 0.3 (or 30%)**

**In simple words**: There’s a **30% chance** that you’ll pick a blue ball first and a red ball second.

---

## **Summary of Key Concepts**:

1. **Marginal Probability**: The chance of **one event** happening.
   - Example: The chance of picking a red ball is **3/5** (or **60%**).

2. **Conditional Probability**: The chance of something happening **after** something else has already happened.
   - Example: After picking a red ball, the chance of picking another red ball is **2/4** (or **50%**).

3. **Joint Probability**: The chance of **two events** happening together.
   - Example: The chance of picking a red ball first and a blue ball second is **3/5 × 2/4 = 6/20 = 30%**.

---

### **Trick to Remember**:

- **Marginal**: One thing happening by itself.
- **Conditional**: One thing happening **after** something else.
- **Joint**: Two things happening **together**.

---

### Quick Recap with Easy Numbers:

1. **Marginal Probability**: Chance of picking a red ball = **3/5**.
2. **Conditional Probability**: Chance of picking a red ball **after** picking one = **2/4**.
3. **Joint Probability**: Chance of picking a red ball first and a blue ball second = **30%**.

```python
# Importing necessary libraries for plotting and numerical calculations
import matplotlib.pyplot as plt   # This library is used for creating visualizations like charts and graphs.
import numpy as np               # This library helps with handling arrays and mathematical operations, though we don't use it much here.

# Define a function to calculate **marginal probability** (chance of one event happening by itself)
def marginal_probability(event_count, total_count):
    """Calculates marginal probability = (event_count / total_count)"""
    # We divide the number of successful outcomes (event_count) by the total number of possible outcomes (total_count)
    return event_count / total_count

# Define a function to calculate **conditional probability** (chance of an event happening given that something else has already happened)
def conditional_probability(event_count_given_condition, total_count_after_condition):
    """Calculates conditional probability = (event_count_given_condition / total_count_after_condition)"""
    # Here we calculate the probability of an event happening after knowing that something else has already happened.
    return event_count_given_condition / total_count_after_condition

# Define a function to calculate **joint probability** (chance of two events happening together)
def joint_probability(prob1, prob2):
    """Calculates joint probability = prob1 * prob2"""
    # To find joint probability, we multiply the individual probabilities of both events happening together.
    return prob1 * prob2

# Set up our example of balls in a bag
# Example 1: Bag with 5 balls (3 red, 2 blue)
red_balls = 3         # We have 3 red balls in the bag.
blue_balls = 2        # We have 2 blue balls in the bag.
total_balls = red_balls + blue_balls  # Total balls in the bag are red + blue = 5 balls.

# Marginal Probability Example 1:
# What is the chance of picking a red ball from the bag?
marginal_red_probability = marginal_probability(red_balls, total_balls)
# We call the marginal_probability function to calculate the chance of picking a red ball (3 red balls out of 5 total balls)

# Display the result of the marginal probability for picking a red ball
print("Marginal Probability of picking a red ball: ", marginal_red_probability)  # Expected output: 3/5 = 0.6 or 60%

# Conditional Probability Example 1:
# Suppose you already picked a red ball, so now there are 2 red balls left and 4 balls in total
remaining_red_balls = 2     # After picking one red ball, 2 red balls are left.
remaining_balls = total_balls - 1  # After picking a ball, 4 balls are left in the bag (total 5 - 1 = 4).

# Now calculate the **conditional probability** of picking a red ball again after one red ball has already been taken out
conditional_red_probability = conditional_probability(remaining_red_balls, remaining_balls)

# Display the result of the conditional probability
print("Conditional Probability of picking a red ball after one red ball is already picked: ", conditional_red_probability)
# Expected output: 2/4 = 0.5 or 50%

# Joint Probability Example 1:
# What is the chance of picking a red ball first and then a blue ball second?
# Step 1: Find the probability of picking a red ball first (3 red balls out of 5 total balls)
prob_red_first = marginal_probability(red_balls, total_balls)  # Using the marginal probability function to calculate the chance of picking a red ball first.

# Step 2: After picking a red ball, there are now 2 blue balls left and 4 balls total
prob_blue_second = marginal_probability(blue_balls, remaining_balls)  # Using the marginal probability function to calculate the chance of picking a blue ball second.

# Now calculate **joint probability** (red ball first and blue ball second)
joint_prob = joint_probability(prob_red_first, prob_blue_second)

# Display the result of the joint probability
print("Joint Probability of picking a red ball first and a blue ball second: ", joint_prob)  # Expected output: (3/5) * (2/4) = 0.3 or 30%

# --- Visualization ---

# Plotting Marginal Probabilities
labels = ['Red Ball', 'Blue Ball']  # Labels for the pie chart
probabilities = [marginal_red_probability, 1 - marginal_red_probability]  # The probabilities of picking a red ball and a blue ball are complementary (1 - prob of red = prob of blue)

# Create a pie chart to visualize the marginal probability
plt.figure(figsize=(6, 6))  # Setting the size of the pie chart
plt.pie(probabilities, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff3366', '#a4fbe8'])
# Plotting the pie chart with:
# - 'probabilities' as the data (the chance of picking a red or blue ball),
# - 'labels' showing "Red Ball" and "Blue Ball",
# - 'autopct' to show percentages in the chart (e.g., 60% or 40%),
# - 'startangle=90' to start the chart from the top,
# - 'colors' to color the slices red and blue.
plt.title("Marginal Probability of Picking a Red or Blue Ball")  # Adding a title to the pie chart
plt.show()  # Displaying the pie chart

# Plotting Conditional Probabilities (Before and After Picking a Red Ball)
fig, ax = plt.subplots(figsize=(8, 6))  # Creating a bar chart with a specific size
ax.bar(['Before Picking Red Ball', 'After Picking Red Ball'], [marginal_red_probability, conditional_red_probability], color=['#ff3366', '#a4fbe8'])
# Creating a bar chart with:
# - The x-axis representing "Before Picking Red Ball" and "After Picking Red Ball",
# - The y-axis representing the probabilities calculated before and after picking a red ball,
# - The bars are colored red and blue.
ax.set_title("Before and After Picking a Red Ball")  # Adding a title to the bar chart
ax.set_ylabel("Probability")  # Adding a label to the y-axis
plt.show()  # Displaying the bar chart

# Plotting Joint Probability (Red First and Blue Second)
plt.figure(figsize=(6, 6))  # Creating another figure for the joint probability bar chart
ax = plt.bar(["Red First and Blue Second"], [joint_prob], color='purple')  # Creating a bar chart with one bar for joint probability

# Fix: Get the current axes to set title and labels properly
ax = plt.gca()  # Get the current axes object so we can modify it (set title, labels, etc.)

# Set title and label correctly now that we have the right axes object
ax.set_title("Joint Probability of Red First and Blue Second")  # Adding a title to the joint probability bar chart
ax.set_ylabel("Probability")  # Adding a label to the y-axis
plt.show()  # Displaying the bar chart
```
